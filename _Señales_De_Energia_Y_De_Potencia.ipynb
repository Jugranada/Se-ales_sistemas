# Señales de energía y potencia

Las señales pueden clasificarse como de energía o de potencia.

Los conceptos de energía y potencia son empleados para la generalización de métricas que permitirán tomar decisiones de convergencia en distintos tipos de análisis en espacios $L^{2}$.

---



## Señales de energía

Se dice que una señal es de energía, si su energía total $E_x$ converge, es decir:

$$
0< E_x < \infty. \tag{1}
$$

Para el caso de una señal $x(t)$ contínua, su energía total es calculada como:

$$
E_x = \|x(t) \|_2^2 = \int_{-\infty}^{\infty} |x(t)|^{2}~dt,  \tag{2}
$$
donde la [norma-2](https://en.wikipedia.org/wiki/Norm_(mathematics)) de la señal $x(t)$ se denota por: $\|x(t) \|_2^2.$ En el mismo sentido, la energía total de la función discreta $x[n]$, es:

$$
E_x = \|x[n] \|_2^2 = \sum_{n=-\infty}^{\infty} |x[n]|^{2}  \tag{3}
$$

Tomando en cuenta que la energía de una señal en un intervalo de tiempo finito $T$ es:

$$
E_T = \int_T |x(t)|^{2}~dt. \tag{4}
$$

La energía total de la señal puede expresarse en terminos del siguiente límite:

$$
E_x = \lim_{T \to \infty} E_T \tag{5}
$$

Para las señales discretas,

$$
E_N = \sum_{n=-N}^{N} |x[n]|^{2} \tag{6}
$$

Y la energía total de $x[n]$ es:

$$
E_x = \lim_{N \to \infty} E_N \tag{7}
$$

## Señales de potencia

Por otro lado, una señal se clasifica como de potencia, si su potencia media $\bar{P}_x$ converge, esto es:

$$
0< \bar{P}_x < \infty \tag{8}
$$

Para el caso de una señal $x(t)$  continua, su potencia media se calcula a partir de la siguiente expresión.

$$
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_T |x(t)|^{2}~dt  \tag{9}
$$

Si $x(t)$ resulta ser periódica, la ecuación anterior puede reescribirse de la siguiente manera:

$$
\bar{P}_x = \lim_{k \to \infty} \frac{1}{k~T_0}~k \int_0^{T_0} |x(t)|^{2}~dt \tag{10}
$$

Donde $T_0$ es el periodo fundamental de la señal, con lo cual se logra hacer una analogía de la ecuación inicial de potencia llevando las repeticiones al infinito, obteniendo el siguiente resultado:

$$
\bar{P}_x = \frac{1}{T_0} \int_0^{T_0} |x(t)|^{2}~dt \tag{11}
$$

Para el caso de una señal discreta $x[n]$, su potencia media está definida por:

$$
\bar{P}_x = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} |x[n]|^{2}  \tag{12}
$$

De igual manera, si $x[n]$ resulta ser periódica, su potencia media estará dada por:

$$
\bar{P}_ x = \frac{1}{N_0} \sum_{n=0}^{N_0-1} |x_[n]|^{2} \tag{13}
$$

Donde $N_0$ es el número mínimo de muestras para el cual la señal $x[n]$ se repite.

La potencia media también se puede expresar en términos de la energía de la señal de la siguiente forma:

* Caso continuo

$$
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T}~E_T \tag{14}
$$

* Caso discreto

$$
\bar{P}_x = \lim_{N \to \infty} \frac{1}{2N+1}~E_N \tag{15}
$$

De lo anterior se puede concluir que:

**-**Si la energía de la señal es finita, entonces su potencia es cero.

**-**Si la señal tiene energía infinita, su potencia puede ser finita o infinita.

**-**Si la potencia de la señal es finita, su energía es infinita.

La potencia instantánea $p(t)$ para el elemento resistivo está dada por la siguiente ecuación:

$$
p(t) = v(t) \cdot i(t) = R \cdot i(t)^{2}
$$

Evaluando en $R = 1 \Omega$,

$$
p(t) = i(t)^{2}
$$

La potencia media de la señal de corriente en un intervalo de tiempo $T$ es entonces la siguiente:

$$
\bar{P}_i = \frac{1}{T} \int_0^{T} i(t)^{2}~dt
$$

Para el caso en el que $i(t)$ coincide con una señal sinusoidal,

$$
\bar{P}_i = \frac{1}{T_0} \int_0^{T_0} \left( I_{\max}~\sin(w_0t+\theta) \right)^{2}~dt\\
\bar{P}_i = \frac{I_{\max}^{2}}{2}
$$

Lo que es equivalente al cuadrado del [valor eficaz](https://en.wikipedia.org/wiki/Root_mean_square) de la señal.

$$
I_{RMS} = \frac{I_{\max}}{\sqrt{2}}
$$

````
from sympy import * #importar todo sympy
init_printing()
t, I_max = symbols('t, I_max', real = True)
w = symbols('w', real = True, nonzero = True)
T0 = 2*pi/w
x = I_max*sin(w*t)
P = integrate(x**2, (t, 0, T0)) /T0
print('La potencia media de la señal es:')
P
````

## Ejercicio 1

Graficar la potencia media de una impedancia $R+jwL$, cuándo es excitada por una fuente sinusoidal.

Graficar la potencia media de una impedancia $R-j\frac{1}{wC} $, cuándo es excitada por una fuente sinusoidal.

Ver cuaderno [https://github.com/amalvarezme/SenalesSistemas/blob/master/1_IntroduccionSyS/6_Potencia_Circutios.ipynb](https://github.com/amalvarezme/SenalesSistemas/blob/master/1_IntroduccionSyS/6_Potencia_Circutios.ipynb) para repaso conceptos básicos potencia en circuitos.

La potencia entre dos fasores $V$ e $I$ se puede calcular mediante la potencia compleja, definida como:

$$
\vec{P} = V \cdot I^*
$$

Donde $\vec{P}$ es la potencia compleja, $V$ es el fasor de tensión y $I^*$ es el conjugado del fasor de corriente. Esta expresión permite conocer tanto la potencia activa (media disipada) como la reactiva (intercambiada con los elementos inductivos o capacitivos del circuito).

Para comprobar la potencia media real, también se puede calcular a partir de la definición temporal como:

$$
P_{\text{media}} = \frac{1}{T} \int_0^T v(t) \cdot i(t) \, dt
$$

donde $v(t)$ e $i(t)$ son las formas de onda de tensión y corriente reconstruidas a partir de los fasores, y $T$ es el periodo de la señal.





````
import numpy as np
import matplotlib.pyplot as plt
import cmath

# --- PARÁMETROS DEL FASOR ---
I = 2                               # Corriente (real, sin ángulo)
Z = 4 + 3j                          # Impedancia compleja (RL)
V = Z * I                           # Tensión compleja
P = V * np.conjugate(I)             # Potencia compleja: V · I*

# --- CONDICIONES PARA SEÑALES EN EL TIEMPO ---
I_max = abs(I)
V_max = abs(V)
theta_I = cmath.phase(I)
theta_V = cmath.phase(V)

w = 300                             # Frecuencia angular
T = 2 * np.pi / w                   # Periodo
t = np.linspace(0, T, 1000)

# --- POTENCIA INSTANTÁNEA Y MEDIA POR INTEGRACIÓN ---
i_t = I_max * np.sin(w * t + theta_I)
v_t = V_max * np.sin(w * t + theta_V)
p_t = v_t * i_t
P_media_integrada = np.trapz(p_t, t) / T

# --- GRAFICAR EN CÍRCULO POLAR ---
N = 2

# Potencia compleja (fasor)
Pr = np.linspace(0, abs(P), N)
Ptheta = cmath.phase(P) * np.ones(N)

# Potencia media por integración (ángulo 0)
Prm = np.linspace(0, P_media_integrada, N)
Pthetam = np.zeros(N)

# Gráfico polar
ax = plt.subplot(111, projection='polar')
ax.plot(Ptheta, Pr, label='Potencia compleja |P|', color='red')
ax.plot(Pthetam, Prm, label='Potencia media (integrada)', color='blue')
ax.legend()
plt.title("Fasor de potencia y potencia media por integración")
plt.show()

# Resultados numéricos
print("Potencia compleja: ", P)
print("Potencia media (por integración):", P_media_integrada)
````

###  Conclusión – Impedancia inductiva $Z = 4 + j3$

- El fasor de potencia $\vec{P}$ forma un ángulo **positivo** respecto al eje real, indicando que **la tensión se adelanta a la corriente**.
- Esto es característico de una **carga inductiva**.

````
import numpy as np
import matplotlib.pyplot as plt
import cmath

# --- PARÁMETROS DEL FASOR (CAPACITOR) ---
I = 2                               # Corriente (real, sin ángulo)
C = 0.01                            # Capacitancia en Faradios
w = 300                             # Frecuencia angular (rad/s)
Z = 4 - 1j / (w * C)                # Impedancia capacitiva: R - j/(wC)
V = Z * I                           # Tensión compleja
P = V * np.conjugate(I)             # Potencia compleja: V · I*

# --- CONDICIONES TEMPORALES ---
I_max = abs(I)
V_max = abs(V)
theta_I = cmath.phase(I)
theta_V = cmath.phase(V)

T = 2 * np.pi / w                   # Periodo
t = np.linspace(0, T, 1000)

# --- POTENCIA INSTANTÁNEA Y MEDIA POR INTEGRACIÓN ---
i_t = I_max * np.sin(w * t + theta_I)
v_t = V_max * np.sin(w * t + theta_V)
p_t = v_t * i_t
P_media_integrada = np.trapz(p_t, t) / T

# --- GRAFICAR EN CÍRCULO POLAR ---
N = 2

# Potencia compleja (fasor)
Pr = np.linspace(0, abs(P), N)
Ptheta = cmath.phase(P) * np.ones(N)

# Potencia media integrada (ángulo 0)
Prm = np.linspace(0, P_media_integrada, N)
Pthetam = np.zeros(N)

# Gráfico polar
ax = plt.subplot(111, projection='polar')
ax.plot(Ptheta, Pr, label='Potencia compleja |P|', color='red')
ax.plot(Pthetam, Prm, label='Potencia media (integrada)', color='blue')
ax.legend()
plt.title("Fasor de potencia y potencia media (Z = 4 - j/(wC))")
plt.show()

# Resultados
print("Potencia compleja: ", P)
print("Potencia media (por integración):", P_media_integrada)
````

###  Conclusión – Impedancia capacitiva $Z = 4 - j3$

- El fasor de potencia $\vec{P}$ tiene un ángulo **negativo**, indicando que **la corriente se adelanta a la tensión**, como ocurre con cargas **capacitivas**.

## Ejercicio 2

Discuta el concepto de valor RMS de una señal a partir de su potencia media. [Ver conceptos RMS](https://en.wikipedia.org/wiki/Root_mean_square).

El **valor RMS** (Root Mean Square, o raíz cuadrática media) de una señal $x(t)$ es una medida útil para representar una señal de forma equivalente a una constante que entrega la misma potencia.

Matemáticamente, el valor RMS de una señal continua en el tiempo está definido como:

$$
x_{\text{RMS}} = \sqrt{\frac{1}{T} \int_0^T x^2(t)\,dt}
$$

Esta definición surge directamente del cálculo de la **potencia media** de una señal. Recordemos que la **potencia media** de una señal $x(t)$ en un intervalo $T$ es:

$$
P = \frac{1}{T} \int_0^T x^2(t)\,dt
$$

Comparando ambas expresiones, se puede ver que:

$$
x_{\text{RMS}} = \sqrt{P}
$$

Es decir, el valor RMS de una señal es la **raíz cuadrada de su potencia media**. Por esta razón, el valor RMS tiene una interpretación física muy importante: es el valor equivalente de una señal variable que produciría la **misma cantidad de potencia** que una señal constante igual a $x_{\text{RMS}}$.

---

### 📌 Conclusión

- El **valor RMS** es una forma de cuantificar la **magnitud efectiva** de una señal.
- Se define directamente a partir de la **potencia media**, lo que le da un significado físico concreto: representa el valor constante equivalente que **produce la misma potencia media**.
- Este concepto es especialmente importante en señales periódicas, como senoidales en circuitos eléctricos.

P = \lim_{T \to \infty} \frac{1}{T} \int_{-T/2}^{T/2} |x(t)|^2 \,dt

##Ejercicio 4

La distancia entre dos señales $x_1(t)$ y $x_2(t)$ se puede dar en términos de la potencia media de la diferencia entre ellas; es decir, la potencia media del error.

$$
d(x_1, x_2) = \bar{P}_{x_1-x_2}=\lim_{T \to \infty} \frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}~dt
$$

Sea $x_1(t)$ y $x_2(t$) dos señales periódicas de periodo fundamental $T_0$ como se muestra a continuación.

$$
x_1(t) = \cos(w_0t), ~~~ w_0 = \frac{2\pi}{T_0}\\
x_2(t)= \left\{ \begin{array}{lcc}
             ~~~1 &   si  & 0 \leq t < \frac{T}{4} \\
             \\ -1 &  si &  \frac{T}{4} \leq t < \frac{3T}{4} \\
             \\ ~~~1 &  si  & \frac{3T}{4} \leq t < T
             \end{array}
   \right.
$$

Determine la distancia entre las dos señales.

---
Para calcular la distancia entre ambas señales, se construyó el **error** $e(t) = x_1(t) - x_2(t)$, se elevó al cuadrado $e^2(t)$, y se integró por tramos en un periodo.

Finalmente, se obtuvo la **potencia media del error** dividiendo la energía del error entre $T$.


````
from sympy import symbols, cos, pi, integrate, simplify  # Importar funciones necesarias de sympy

# Declarar variables simbólicas
t, T, A = symbols('t T A', real=True, positive=True)  # t = tiempo, T = periodo, A = amplitud

# Definir la frecuencia angular w0 = 2π / T
w0 = 2 * pi / T

# Definir la señal x1(t) = A * cos(w0 * t)
x1 = A * cos(w0 * t)

# Tramo 1: x2(t) = 1  → e(t) = x1(t) - 1
e1 = (x1 - 1)**2
int1 = integrate(e1, (t, 0, T/4))  # Integrar el primer tramo de 0 a T/4

# Tramo 2: x2(t) = -1 → e(t) = x1(t) + 1
e2 = (x1 + 1)**2
int2 = integrate(e2, (t, T/4, 3*T/4))  # Integrar el segundo tramo de T/4 a 3T/4

# Tramo 3: x2(t) = 1  → e(t) = x1(t) - 1
e3 = (x1 - 1)**2
int3 = integrate(e3, (t, 3*T/4, T))  # Integrar el tercer tramo de 3T/4 a T

# Sumar todas las integrales
E_total = int1 + int2 + int3

# Dividir por T para obtener la potencia media
Pe = simplify(E_total / T)

# Mostrar resultado
print('La potencia media del error es:')
Pe
````

---

### Conclusiones

- La potencia media del error $\bar{P}_e$ representa la **distancia cuadrática promedio** entre las señales $x_1(t)$ y $x_2(t)$ en un periodo.
- Esta medida permite cuantificar **qué tan diferentes son dos señales periódicas** desde el punto de vista energético.
- El resultado obtenido permite **evaluar qué tanto se aleja una forma de onda de referencia (como un coseno) de una aproximación por tramos (como una señal cuadrada)**.
- Este tipo de análisis es fundamental en contextos como compresión de señales, aproximación por series de Fourier y control.

## Ejercicio 5

Demostrar que la norma al cuadrado de una señal equivale a su energía.

### Demostración: la norma al cuadrado de una señal equivale a su energía

Para comprobar que la **norma al cuadrado** de una señal equivale a su **energía total**, se utiliza como ejemplo la señal:

$$
x(t) = e^{-|t|}
$$

Esta señal es continua, simétrica, decrece exponencialmente, y tiene energía finita.

---

### Definiciones teóricas

- La **energía total** de una señal se define como:

$$
E_x = \int_{-\infty}^{\infty} |x(t)|^2 \, dt
$$

- La **norma 2** de una señal es:

$$
\|x(t)\| = \sqrt{ \int_{-\infty}^{\infty} |x(t)|^2 \, dt }
$$

Por lo tanto:

$$
\|x(t)\|^2 = E_x
$$

---

###  Demostración simbólica en Python

Se utilizó `SymPy` para definir la señal $x(t) = e^{-|t|}$ y calcular:

1. La **energía total** de la señal:
$$
E_x = \int_{-\infty}^{\infty} e^{-2|t|} \, dt = 1
$$

2. La **norma al cuadrado**:
$$
\|x(t)\|^2 = \left( \sqrt{ \int_{-\infty}^{\infty} e^{-2|t|} \, dt } \right)^2 = 1
$$

Ambos resultados coinciden, lo que confirma la igualdad.

---


````
from sympy import symbols, exp, integrate, sqrt, Abs, oo, simplify, init_printing

# Inicializar impresión bonita
init_printing()

# Declarar variable simbólica
t = symbols('t', real=True)

# Definir la señal
x = exp(-Abs(t))  # x(t) = e^{-|t|}

# Calcular energía de la señal
energia = integrate(x**2, (t, -oo, oo))

# Calcular norma
norma = sqrt(integrate(x**2, (t, -oo, oo)))

# Cuadrado de la norma
norma_cuadrado = norma**2

# Mostrar resultados
print("Energía total de la señal:")
display(simplify(energia))

print("\nNorma al cuadrado de la señal:")
display(simplify(norma_cuadrado))
````

### Conclusión

- Se demuestra que para señales cuadráticamente integrables, la **norma al cuadrado** de la señal **es exactamente igual** a su **energía total**.
- Esta propiedad es fundamental en análisis de señales y sistemas, ya que permite usar el concepto de norma como medida de “magnitud” o “distancia energética” de una señal.
