### A. Serie de Fourier: exponencial, trigonom√©trica y compacta

La Serie de Fourier permite representar una se√±al peri√≥dica en el tiempo continuo como una suma infinita de funciones arm√≥nicas (senos y cosenos, o exponenciales complejas).

#### Serie de Fourier trigonom√©trica

Representa una se√±al peri√≥dica x(t) como una suma de senos y cosenos:

$$
x(t) = a_0 + \sum_{n=1}^{\infty} \left[a_n \cos(n \omega_0 t) + b_n \sin(n \omega_0 t)\right]
$$

- Utiliza funciones reales.
- Los coeficientes \( a_n \) y \( b_n \) son reales.
- Es la forma m√°s intuitiva, con interpretaci√≥n directa de amplitud y fase.

#### Serie de Fourier exponencial

Usa funciones exponenciales complejas para representar la se√±al:

$$
x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j n \omega_0 t}
$$

- Los coeficientes \( c_n \) son generalmente complejos.
- Relaciona directamente con la transformada de Fourier.
- Permite an√°lisis m√°s compacto en el dominio complejo.

Relaci√≥n con los coeficientes trigonom√©tricos:

$$
c_n = \frac{1}{2}(a_n - j b_n), \quad c_{-n} = \frac{1}{2}(a_n + j b_n)
$$

#### Serie de Fourier compacta

Se refiere a una notaci√≥n simplificada que unifica la expresi√≥n en una sola f√≥rmula (generalmente la forma exponencial), para expresar de forma concisa la representaci√≥n espectral.

- Suele usarse cuando se estudia el espectro como conjunto de coeficientes \( c_n \).
- Simplifica el an√°lisis algebraico.

---

#### Comparaci√≥n

| Caracter√≠stica                   | Trigonom√©trica                          | Exponencial                                | Compacta                |
|----------------------------------|------------------------------------------|---------------------------------------------|--------------------------|
| Base                             | Seno y coseno                            | Exponenciales complejas                     | Id√©ntica a la exponencial |
| Coeficientes                     | Reales                                   | Complejos                                   | Complejos                |
| Simetr√≠a                         | Expl√≠cita (senos impares, cosenos pares) | Impl√≠cita (incluye positivos y negativos)   | Impl√≠cita                |
| Uso com√∫n                        | An√°lisis f√≠sico e intuitivo              | Teor√≠a de se√±ales, sistemas y comunicaciones | Matem√°tica/te√≥rica       |
| Relaci√≥n con TF y DFT            | Indirecta                                | Directa                                     | Directa                  |




### B. Comparaci√≥n entre Serie de Fourier, Transformada de Fourier, DTFT y DFT

A continuaci√≥n se comparan distintas herramientas de an√°lisis espectral que se aplican seg√∫n si la se√±al es continua o discreta, peri√≥dica o no peri√≥dica.

---

####  Transformada de Fourier (TF)

La Transformada de Fourier generaliza la Serie de Fourier para se√±ales no peri√≥dicas en tiempo continuo. Transforma una se√±al x(t) en su representaci√≥n espectral continua X(œâ):

$$
X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j \omega t} \, dt
$$

- Entrada: se√±al en tiempo continuo, no peri√≥dica.
- Salida: espectro de frecuencia continuo.
- Aplicaciones: procesamiento de se√±ales anal√≥gicas, comunicaciones, sistemas lineales.

---

####  Transformada de Fourier en Tiempo Discreto (DTFT)

Es la versi√≥n de la TF para se√±ales discretas (muestras infinitas). Representa la se√±al en el dominio de la frecuencia continua:

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n}
$$

- Entrada: se√±al en tiempo discreto, no peri√≥dica.
- Salida: espectro de frecuencia continuo (pero peri√≥dico en 2œÄ).
- Aplicaciones: an√°lisis te√≥rico de sistemas digitales, filtros.

---

####  Transformada Discreta de Fourier (DFT)

La DFT representa una secuencia discreta y finita en t√©rminos de un n√∫mero finito de componentes de frecuencia:

$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} k n}, \quad k = 0,1,\dots,N-1
$$

- Entrada: se√±al discreta, de duraci√≥n finita (N muestras).
- Salida: espectro discreto, finito y peri√≥dico.
- Aplicaciones: implementaciones pr√°cticas (FFT), an√°lisis num√©rico.

---

###  Comparaci√≥n General

| Transformaci√≥n        | Tipo de se√±al de entrada        | Tipo de espectro         | Periodicidad entrada | Periodicidad salida |
|-----------------------|----------------------------------|---------------------------|-----------------------|----------------------|
| Serie de Fourier      | Continua, peri√≥dica              | Discreto                  | Peri√≥dica             | Discreta             |
| TF                    | Continua, no peri√≥dica           | Continua                  | No peri√≥dica          | Continua             |
| DTFT                  | Discreta, no peri√≥dica (infinita)| Continua y peri√≥dica      | No peri√≥dica          | Peri√≥dica (2œÄ)       |
| DFT                   | Discreta, finita (N muestras)    | Discreta y peri√≥dica      | Finita                | Discreta y peri√≥dica |

---

Estas herramientas permiten analizar se√±ales en diferentes dominios y contextos. La elecci√≥n adecuada depende del tipo de se√±al y del an√°lisis que se desea realizar: an√°lisis te√≥rico (TF, DTFT) o computacional (DFT).

### D. ¬øEn qu√© consiste el algoritmo FFT y cu√°l es su utilidad?

####  ¬øQu√© es la FFT?

La Fast Fourier Transform (FFT) es un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT) de una se√±al. La DFT convierte una secuencia de valores en el dominio del tiempo en una representaci√≥n en el dominio de la frecuencia.

El c√°lculo directo de la DFT tiene una complejidad computacional de orden:

$$
\mathcal{O}(N^2)
$$

donde N es el n√∫mero de muestras. Esto se debe a que se requieren N multiplicaciones complejas para cada una de las N frecuencias.

La FFT reduce dr√°sticamente esta complejidad a:

$$
\mathcal{O}(N \log_2 N)
$$

Este aumento de eficiencia se logra aprovechando simetr√≠as y redundancias en los c√°lculos de la DFT, especialmente cuando N es una potencia de 2.

---

####  ¬øC√≥mo funciona?

El algoritmo FFT m√°s conocido es el de Cooley-Tukey, que descompone la DFT de tama√±o N en DFTs m√°s peque√±as recursivamente, hasta llegar a DFTs de tama√±o 2 (caso base).

Por ejemplo, para una se√±al de longitud N = 8:

- La se√±al se divide en dos grupos: √≠ndices pares e impares.
- Se calcula la FFT para cada grupo recursivamente.
- Luego, se combinan los resultados usando factores de forma conocidos como "twiddle factors" (ra√≠ces de la unidad).

Este enfoque divide el problema grande en problemas m√°s peque√±os (divide y vencer√°s), lo que reduce considerablemente el n√∫mero total de operaciones.

---

####  Utilidad de la FFT

- Permite el an√°lisis espectral en tiempo real o casi real.
- Es la base de aplicaciones como:
  - Procesamiento de audio y m√∫sica.
  - Compresi√≥n de se√±ales.
  - Detecci√≥n de fallos en sistemas el√©ctricos.
  - Reconocimiento de patrones y voz.
  - An√°lisis en comunicaciones (OFDM, modulaci√≥n digital).

---

####  Comparaci√≥n de costos computacionales

| M√©todo | N√∫mero de operaciones complejas | Complejidad |
|--------|------------------------------|-------------|
| DFT directa | N¬≤                       | ùí™(N¬≤)       |
| FFT (Cooley-Tukey) | N log‚ÇÇ N         | ùí™(N log‚ÇÇ N) |

Para N = 1024:

- DFT requiere m√°s de 1 mill√≥n de operaciones.
- FFT reduce ese n√∫mero a aproximadamente 10,240.

Esto permite que transformadas grandes sean computacionalmente viables y r√°pidas, incluso en dispositivos embebidos.



### E. Explicaci√≥n detallada del funcionamiento de la FFT y comparaci√≥n de su costo computacional con la DFT

####  Funcionamiento del algoritmo FFT

La Fast Fourier Transform (FFT) es un algoritmo que calcula de forma eficiente la Transformada Discreta de Fourier (DFT). Su funcionamiento se basa en descomponer la DFT de tama√±o N en subproblemas m√°s peque√±os, aprovechando simetr√≠as y redundancias matem√°ticas.

Uno de los algoritmos m√°s comunes de FFT es el Cooley-Tukey. Este asume que N (longitud de la se√±al) es una potencia de 2 (aunque hay variantes para otros casos) y realiza lo siguiente:

1. Divide la se√±al de entrada en dos secuencias:
   - Los elementos en √≠ndices pares.
   - Los elementos en √≠ndices impares.

2. Calcula recursivamente la FFT de estas dos secuencias de longitud N/2.

3. Usa propiedades de periodicidad y simetr√≠a de las ra√≠ces de la unidad para combinar los resultados de las FFT menores y reconstruir la FFT total.

Esto se repite hasta que las FFTs sean de tama√±o 2 (caso base), lo que permite que el n√∫mero total de operaciones se reduzca significativamente.

Ejemplo visual (para N = 8):

- Entrada x[n]: 8 muestras.
- Se divide en dos grupos: x[0], x[2], x[4], x[6] y x[1], x[3], x[5], x[7].
- Se calcula FFT de cada grupo (4 muestras).
- Luego se combinan los resultados usando factores twiddle:
  W_N^k = e^{-j 2œÄk/N}

---

####  Comparaci√≥n del costo computacional: FFT vs. DFT

| M√©todo            | N√∫mero de operaciones | Complejidad computacional |
|------------------|------------------------|----------------------------|
| DFT directa       | N¬≤ operaciones complejas | ùí™(N¬≤)                      |
| FFT (Cooley-Tukey) | N log‚ÇÇ N operaciones     | ùí™(N log‚ÇÇ N)                |

La diferencia se hace notable para valores grandes de N.

Ejemplo para N = 1024:

- DFT directa: 1024¬≤ = 1,048,576 operaciones.
- FFT: 1024 √ó log‚ÇÇ(1024) = 1024 √ó 10 = 10,240 operaciones.

Esto significa que la FFT es aproximadamente 100 veces m√°s r√°pida que la DFT directa para N = 1024. Esta eficiencia es clave para aplicaciones en tiempo real, como procesamiento de se√±ales, an√°lisis de espectro, comunicaciones y procesamiento de audio e imagen.

---

Conclusi√≥n: la FFT no solo conserva el resultado de la DFT, sino que permite obtenerlo con mucho menor esfuerzo computacional, lo que la convierte en una herramienta esencial en ingenier√≠a y ciencias aplicadas.


### **2. Modulaci√≥n AM por detecci√≥n coherente**

En este ejercicio se explora la modulaci√≥n por amplitud (AM) utilizando detecci√≥n coherente, una t√©cnica ampliamente usada en comunicaciones anal√≥gicas. Se estudian sus fundamentos y aplicaciones. Posteriormente, se implementa en Python un ejemplo ilustrativo en el que se visualizan las se√±ales en el dominio del tiempo y la frecuencia (utilizando la transformada r√°pida de Fourier unidireccional, rfft). ### A. ¬øQu√© es la modulaci√≥n AM con detecci√≥n coherente?

La modulaci√≥n por amplitud (AM, por sus siglas en ingl√©s) es una t√©cnica mediante la cual la amplitud de una se√±al portadora senoidal se var√≠a de forma proporcional a una se√±al de mensaje o informaci√≥n. En la detecci√≥n coherente (tambi√©n llamada detecci√≥n s√≠ncrona), la se√±al modulada es demodulada utilizando una se√±al portadora que est√° perfectamente sincronizada en frecuencia y fase con la portadora original utilizada en la transmisi√≥n.

Esta sincronizaci√≥n permite recuperar con mayor precisi√≥n la se√±al de mensaje original, ya que se evita la distorsi√≥n que puede generarse con otros m√©todos de detecci√≥n no coherente (como la detecci√≥n por envolvente).

#### Se√±al modulada en AM:
La se√±al AM con modulaci√≥n en amplitud se puede expresar como:

$$
s(t) = A_c \left[1 + m \cdot x(t)\right] \cdot \cos(2\pi f_c t)
$$

Donde:
- \( A_c \): amplitud de la portadora.
- \( x(t) \): se√±al de mensaje (normalizada entre -1 y 1).
- \( f_c \): frecuencia de la portadora.
- \( m \): √≠ndice de modulaci√≥n.

Para que la se√±al est√© correctamente modulada, se requiere que \( 0 < m \leq 1 \). Si \( m > 1 \), se produce sobre-modulaci√≥n.

---

### Aplicaciones de la modulaci√≥n AM con detecci√≥n coherente

La detecci√≥n coherente se utiliza cuando se requiere mayor fidelidad en la recuperaci√≥n de la se√±al de mensaje. Algunas aplicaciones comunes incluyen:

- Sistemas de comunicaci√≥n anal√≥gica de alta calidad.
- Comunicaciones en radio AM (en algunos receptores avanzados).
- Comunicaciones en banda base de sistemas embebidos.
- Transmisi√≥n de datos mediante modulaci√≥n digital (por ejemplo, QAM, que requiere detecci√≥n coherente).
- Aplicaciones en radar y telemetr√≠a.

Este tipo de detecci√≥n requiere que tanto el transmisor como el receptor compartan un reloj com√∫n o un sistema de sincronizaci√≥n de portadora.


### B. Ejemplo ilustrativo

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# Par√°metros generales
Fs = 10000  # Frecuencia de muestreo
t = np.arange(0, 0.01, 1/Fs)  # Tiempo de 10 ms

# Par√°metros de modulaci√≥n
Fc = 1000         # Frecuencia de la portadora
Ac = 1            # Amplitud de la portadora
indice_mod = 0.8  # √çndice de modulaci√≥n definido por el usuario

# Se√±al mensaje tipo pulso rectangular
m_pulso = np.where((t >= 0.002) & (t < 0.008), 1.0, 0.0)

# Se√±al mensaje tipo coseno
fm = 200  # frecuencia del mensaje
m_coseno = np.cos(2 * np.pi * fm * t)

# Funci√≥n para modulaci√≥n AM
def am_modular(mensaje, Fc, Fs, Ac, indice_mod):
    m_norm = mensaje / np.max(np.abs(mensaje))  # Normalizar mensaje
    portadora = Ac * np.cos(2 * np.pi * Fc * t)
    y = (1 + indice_mod * m_norm) * portadora
    return y

# Se√±ales moduladas
y_pulso = am_modular(m_pulso, Fc, Fs, Ac, indice_mod)
y_coseno = am_modular(m_coseno, Fc, Fs, Ac, indice_mod)

# Gr√°ficas en el tiempo
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(t, m_pulso)
plt.title('Mensaje: Pulso rectangular')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.subplot(2, 2, 2)
plt.plot(t, y_pulso)
plt.title('AM modulada (pulso)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.subplot(2, 2, 3)
plt.plot(t, m_coseno)
plt.title('Mensaje: Coseno')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.subplot(2, 2, 4)
plt.plot(t, y_coseno)
plt.title('AM modulada (coseno)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()

# Espectros usando rfft
def graficar_espectro(signal, Fs, title):
    N = len(signal)
    freqs = rfftfreq(N, 1/Fs)
    spectrum = np.abs(rfft(signal)) / N
    plt.plot(freqs, spectrum)
    plt.title(title)
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Magnitud')

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
graficar_espectro(y_pulso, Fs, 'Espectro AM (pulso)')

plt.subplot(1, 2, 2)
graficar_espectro(y_coseno, Fs, 'Espectro AM (coseno)')

plt.tight_layout()
plt.show()
````

- Gr√°fica 5 (izquierda): espectro de la se√±al AM con mensaje tipo pulso rectangular. Se observan m√∫ltiples componentes frecuenciales debido a la naturaleza no peri√≥dica del pulso. El espectro es m√°s ancho comparado con el caso del coseno.

Finalmente, se realiza un an√°lisis param√©trico variando R y C, discutiendo c√≥mo estos valores impactan el contenido arm√≥nico y el desempe√±o energ√©tico del circuito.


### a) ¬øEn qu√© consiste la Distorsi√≥n Total de Arm√≥nicos (THD)?

La Distorsi√≥n Total de Arm√≥nicos (THD) es una medida que cuantifica el nivel de distorsi√≥n de una se√±al peri√≥dica debido a la presencia de arm√≥nicos. Los arm√≥nicos son componentes de frecuencia que son m√∫ltiplos enteros de la frecuencia fundamental de una se√±al.

En los sistemas el√©ctricos y electr√≥nicos, especialmente en circuitos con componentes no lineales (como rectificadores, inversores o cargas electr√≥nicas), las se√±ales de corriente o voltaje pueden distorsionarse y alejarse de una forma senoidal pura. Esta distorsi√≥n introduce componentes arm√≥nicas que degradan la calidad de la energ√≠a el√©ctrica y pueden afectar el rendimiento de los equipos.

El THD se define como el cociente entre la ra√≠z cuadrada de la suma de los cuadrados de todas las componentes arm√≥nicas (excepto la fundamental) y la amplitud de la componente fundamental. Se expresa generalmente en porcentaje (%).

$$
\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

Donde:

$$
V_1 = \text{Amplitud de la componente fundamental}
$$

$$
V_2, V_3, \dots, V_n = \text{Amplitudes de los arm√≥nicos de orden 2 al } n
$$

Una **THD baja** indica que la se√±al est√° cerca de una forma senoidal pura. En cambio, una **THD alta** revela una fuerte presencia de arm√≥nicos y, por tanto, mayor distorsi√≥n.

### b) ¬øQu√© es el factor de potencia en un circuito el√©ctrico?

El **factor de potencia (FP)** es una medida que describe cu√°n eficientemente se utiliza la energ√≠a el√©ctrica en un sistema. Representa la relaci√≥n entre la **potencia activa** (realmente utilizada para realizar trabajo √∫til) y la **potencia aparente** (total suministrada por la fuente).

Se define como:

$$
\text{FP} = \frac{P}{S}
$$

donde:

- \( P \): Potencia activa (o real), medida en watts (W).  
- \( S \): Potencia aparente, medida en volt-amperios (VA).

El factor de potencia puede expresarse como un valor entre 0 y 1, o como un porcentaje. Un **FP = 1** (o 100%) indica que toda la potencia suministrada se est√° utilizando eficazmente. Un **FP bajo** significa que una parte significativa de la potencia est√° siendo desperdiciada en forma de potencia reactiva.

En sistemas sinusoidales, el FP tambi√©n se puede interpretar como el **coseno del √°ngulo de desfase \( \phi \)** entre la corriente y el voltaje:

$$
\text{FP} = \cos(\phi)
$$

- Si el FP es **inductivo**, la corriente **retrasa** al voltaje.
- Si es **capacitivo**, la corriente **adelanta** al voltaje.

Un bajo factor de potencia puede aumentar las p√©rdidas en el sistema y reducir la capacidad de transmisi√≥n, por lo que es deseable mantenerlo lo m√°s alto posible.


### c) ¬øC√≥mo puede calcularse el THD usando la Transformada R√°pida de Fourier (FFT)?

La Transformada R√°pida de Fourier (FFT) permite descomponer una se√±al peri√≥dica (como una corriente o voltaje en un circuito) en una suma de componentes senoidales (arm√≥nicos), cada una con una frecuencia, amplitud y fase espec√≠ficas.

La Distorsi√≥n Total Arm√≥nica (THD) puede calcularse utilizando los coeficientes de la FFT de la se√±al. El procedimiento general es:

1. Se calcula la FFT de la se√±al de inter√©s (por ejemplo, la corriente del circuito).
2. Se obtiene el m√≥dulo (magnitud) de la FFT, que corresponde a las amplitudes de cada componente espectral.
3. Se identifica la componente de frecuencia fundamental (la de menor frecuencia distinta de cero).
4. Se suman los cuadrados de las amplitudes de los arm√≥nicos (todas las frecuencias m√∫ltiples de la fundamental, excluyendo la fundamental).
5. Finalmente, se calcula el THD con la f√≥rmula:

$$
\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

donde:
$$
V_1 = \text{Amplitud de la componente fundamental}
$$

$$
V_2, V_3, \dots, V_n = \text{Amplitudes de los arm√≥nicos de orden 2 al } n
$$

### d) ¬øC√≥mo se relaciona el THD con la distorsi√≥n del factor de potencia?

El factor de potencia (FP) en sistemas con se√±ales no sinusoidales se ve afectado por dos tipos de distorsiones:

1. El desfase entre la tensi√≥n y la corriente (factor de potencia de desplazamiento).
2. La presencia de arm√≥nicos en la se√±al de corriente (factor de potencia de distorsi√≥n).

En este contexto, el THD (Distorsi√≥n Total Arm√≥nica) de la corriente se relaciona directamente con el segundo factor: la distorsi√≥n arm√≥nica.

El factor de potencia total se puede calcular como:

$$
\text{FP} = \frac{P}{S} = \frac{V_{\text{rms}} I_1 \cos(\phi)}{V_{\text{rms}} I_{\text{rms}}} = \frac{I_1 \cos(\phi)}{I_{\text{rms}}}
$$

Donde:

$$
\begin{aligned}
I_1 & : \text{Componente fundamental de la corriente (amplitud de la FFT en la frecuencia fundamental)} \\
I_{\text{rms}} & : \text{Valor eficaz total de la corriente (incluye fundamental y arm√≥nicos)} \\
\phi & : \text{√Ångulo de desfase entre la tensi√≥n y la corriente fundamental}
\end{aligned}
$$

Ahora bien, la relaci√≥n entre el THD y el FP se da a trav√©s del factor de potencia de distorsi√≥n, dado por:

$$
\text{FP}_{\text{distorsi√≥n}} = \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

De esta forma, el factor de potencia total se puede expresar como:

$$
\text{FP} = \text{FP}_{\text{desplazamiento}} \times \text{FP}_{\text{distorsi√≥n}} = \cos(\phi) \times \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

Esto significa que incluso si el desfase entre tensi√≥n y corriente fundamental es peque√±o (es decir, \( \cos(\phi) \approx 1 \)), un alto contenido de arm√≥nicos (alto THD) reducir√° significativamente el factor de potencia total.


e) Dise√±ar y simular un rectificador de onda completa en dos casos:

I. Carga netamente resistiva

II. Carga RC en serie:

**I. Carga netamente resistiva**

Se va a simular un rectificador de onda completa alimentado por una se√±al senoidal de 60 Hz con amplitud de 120‚ÄØV, conectada a una carga puramente resistiva.

**Paso 1: Definir par√°metros y generar la se√±al de entrada**

````python
# Par√°metros de simulaci√≥n
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import scipy.fftpack as fft

# Definici√≥n de par√°metros
Fo = 60            # Frecuencia de alimentaci√≥n (Hz)
Fs = 30 * Fo       # Frecuencia de muestreo
To = 1 / Fo        # Periodo fundamental
Ts = 1 / Fs        # Periodo de muestreo
t = np.arange(0, 5 * To, Ts)  # tiempo para 5 periodos

A = 120  # Amplitud de entrada (voltaje)
in_o = A * np.sin(2 * np.pi * Fo * t)  # se√±al senoidal de entrada
rec_c = sig.square(2 * np.pi * Fo * t)  # tren de pulsos para rectificaci√≥n completa

# Se√±al rectificada completamente
in_ = in_o * rec_c

# Graficar se√±al de entrada rectificada
plt.plot(t, in_, label='Entrada rectificada')
plt.xlabel('t [s]')
plt.ylabel('Voltaje [V]')
plt.title('Entrada al circuito: Rectificador onda completa')
plt.legend()
plt.grid()
plt.show()
````

**Paso 2: Calcular y graficar el espectro para calcular el THD**

````python
# FFT de la se√±al rectificada
Xf = np.fft.rfft(in_)
vfre = np.fft.rfftfreq(len(in_), 1/Fs)

# Graficar espectro
plt.stem(vfre, np.abs(Xf)/len(in_))  # sin use_line_collection
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de la corriente rectificada')
plt.grid()
plt.show()
````

**Paso 3: C√°lculo del THD y del factor de potencia por distorsi√≥n**

````python
# THD desde componentes espectrales
Xmag = np.abs(Xf)/len(in_)
fundamental = Xmag[np.argmax((vfre >= Fo) & (vfre <= Fo+1))]  # componente en 60 Hz
harmonics = Xmag[(vfre > Fo + 1)]  # arm√≥nicos superiores (excluye la fundamental)

# Calcular THD
THD = np.sqrt(np.sum(harmonics**2)) / fundamental
print(f"THD: {THD:.4f} ({THD*100:.2f}%)")

# Calcular factor de potencia por distorsi√≥n
FP_dist = 1 / np.sqrt(1 + THD**2)
print(f"Factor de potencia (por distorsi√≥n): {FP_dist:.4f}")
````

**II. Carga RC en serie:**

````python
# Par√°metros del circuito
f = 60              # Frecuencia de red [Hz]
w = 2*np.pi*f       # Frecuencia angular [rad/s]
Vm = 170            # Voltaje pico (120 Vrms ‚âà 170 V pico)
R = 100             # Resistencia [Ohm]
C = 100e-6          # Capacitancia [F]

T = 1/f             # Periodo de la se√±al
tf = 4*T            # Tiempo total de simulaci√≥n
n = 1000            # Puntos por ciclo
t = np.linspace(0, tf, int(n*tf/T))
dt = t[1] - t[0]

# Voltaje de entrada (onda completa rectificada)
v_in = Vm*np.abs(np.sin(w*t))

# Inicializaci√≥n de variables
i_rc = np.zeros(len(t))
v_c = np.zeros(len(t))

# Simulaci√≥n de corriente y voltaje en el capacitor
for k in range(1, len(t)):
    di = (v_in[k] - R*i_rc[k-1] - v_c[k-1]) / R
    i_rc[k] = i_rc[k-1] + di*dt
    v_c[k] = v_c[k-1] + i_rc[k]*dt/C
````

### Forma de onda de corriente y voltaje en el condensador

````python
plt.figure(figsize=(10,5))
plt.plot(t, v_in, label='$v_{in}(t)$', linestyle='--', color='gray')
plt.plot(t, i_rc, label='$i_{RC}(t)$', color='green')
plt.plot(t, v_c, label='$v_C(t)$', color='red')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Rectificador de onda completa con carga RC en serie')
plt.legend()
plt.grid()
plt.show()
````

### Espectro de la corriente con carga RC


````python
from scipy.fft import fft, fftfreq

Xf_rc = fft(i_rc)
vfre_rc = fftfreq(len(i_rc), dt)

plt.figure(figsize=(10,4))
plt.stem(vfre_rc, np.abs(Xf_rc)/len(i_rc))
plt.xlim(0, 1000)
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de $i_{RC}(t)$')
plt.grid()
plt.show()
````

### C√°lculo del THD y del factor de potencia por distorsi√≥n

````python
# Magnitud del espectro
mag_rc = np.abs(Xf_rc[:len(Xf_rc)//2]) / len(i_rc)

# Componente fundamental (60 Hz)
V1_rc = mag_rc[np.argmin(np.abs(vfre_rc[:len(mag_rc)] - f))]

# Arm√≥nicos relevantes
harmonicos = [2*f, 3*f, 4*f, 5*f, 6*f]
sum_h_rc = 0
for h in harmonicos:
    idx = np.argmin(np.abs(vfre_rc[:len(mag_rc)] - h))
    sum_h_rc += mag_rc[idx]**2

# THD
THD_rc = np.sqrt(sum_h_rc) / V1_rc
print(f"THD (RC): {THD_rc:.4f} ({THD_rc*100:.2f}%)")

# Factor de potencia por distorsi√≥n
FPd_rc = 1 / np.sqrt(1 + THD_rc**2)
print(f"Factor de potencia por distorsi√≥n (RC): {FPd_rc:.4f}")
````

### Conclusi√≥n

Se observa que la presencia del condensador modifica la forma de onda de la corriente, introduciendo componentes adicionales en el espectro (m√°s arm√≥nicos). Esto provoca un incremento en el THD y una disminuci√≥n en el factor de potencia por distorsi√≥n en comparaci√≥n con la carga puramente resistiva.

Este an√°lisis es fundamental para entender c√≥mo la naturaleza de la carga afecta la calidad de la corriente consumida y su impacto en la eficiencia energ√©tica del sistema.
