En este ejercicio se analizan las principales herramientas matemáticas utilizadas para representar y analizar señales en los dominios del tiempo y la frecuencia. ### A. Serie de Fourier: exponencial, trigonométrica y compacta

La Serie de Fourier permite representar una señal periódica en el tiempo continuo como una suma infinita de funciones armónicas (senos y cosenos, o exponenciales complejas).

#### Serie de Fourier trigonométrica

Representa una señal periódica x(t) como una suma de senos y cosenos:

$$
x(t) = a_0 + \sum_{n=1}^{\infty} \left[a_n \cos(n \omega_0 t) + b_n \sin(n \omega_0 t)\right]
$$

- Utiliza funciones reales.
- Los coeficientes \( a_n \) y \( b_n \) son reales.
- Es la forma más intuitiva, con interpretación directa de amplitud y fase.

#### Serie de Fourier exponencial

Usa funciones exponenciales complejas para representar la señal:

$$
x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j n \omega_0 t}
$$

- Los coeficientes \( c_n \) son generalmente complejos.
- Relaciona directamente con la transformada de Fourier.
- Permite análisis más compacto en el dominio complejo.

Relación con los coeficientes trigonométricos:

$$
c_n = \frac{1}{2}(a_n - j b_n), \quad c_{-n} = \frac{1}{2}(a_n + j b_n)
$$

#### Serie de Fourier compacta

Se refiere a una notación simplificada que unifica la expresión en una sola fórmula (generalmente la forma exponencial), para expresar de forma concisa la representación espectral.

- Suele usarse cuando se estudia el espectro como conjunto de coeficientes \( c_n \).
- Simplifica el análisis algebraico.

---

#### Comparación

| Característica                   | Trigonométrica                          | Exponencial                                | Compacta                |
|----------------------------------|------------------------------------------|---------------------------------------------|--------------------------|
| Base                             | Seno y coseno                            | Exponenciales complejas                     | Idéntica a la exponencial |
| Coeficientes                     | Reales                                   | Complejos                                   | Complejos                |
| Simetría                         | Explícita (senos impares, cosenos pares) | Implícita (incluye positivos y negativos)   | Implícita                |
| Uso común                        | Análisis físico e intuitivo              | Teoría de señales, sistemas y comunicaciones | Matemática/teórica       |
| Relación con TF y DFT            | Indirecta                                | Directa                                     | Directa                  |




### B. Comparación entre Serie de Fourier, Transformada de Fourier, DTFT y DFT

A continuación se comparan distintas herramientas de análisis espectral que se aplican según si la señal es continua o discreta, periódica o no periódica.

---

####  Transformada de Fourier (TF)

La Transformada de Fourier generaliza la Serie de Fourier para señales no periódicas en tiempo continuo. Transforma una señal x(t) en su representación espectral continua X(ω):

$$
X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j \omega t} \, dt
$$

- Entrada: señal en tiempo continuo, no periódica.
- Salida: espectro de frecuencia continuo.
- Aplicaciones: procesamiento de señales analógicas, comunicaciones, sistemas lineales.

---

####  Transformada de Fourier en Tiempo Discreto (DTFT)

Es la versión de la TF para señales discretas (muestras infinitas). Representa la señal en el dominio de la frecuencia continua:

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n}
$$

- Entrada: señal en tiempo discreto, no periódica.
- Salida: espectro de frecuencia continuo (pero periódico en 2π).
- Aplicaciones: análisis teórico de sistemas digitales, filtros.

---

####  Transformada Discreta de Fourier (DFT)

La DFT representa una secuencia discreta y finita en términos de un número finito de componentes de frecuencia:

$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} k n}, \quad k = 0,1,\dots,N-1
$$

- Entrada: señal discreta, de duración finita (N muestras).
- Salida: espectro discreto, finito y periódico.
- Aplicaciones: implementaciones prácticas (FFT), análisis numérico.

---

###  Comparación General

| Transformación        | Tipo de señal de entrada        | Tipo de espectro         | Periodicidad entrada | Periodicidad salida |
|-----------------------|----------------------------------|---------------------------|-----------------------|----------------------|
| Serie de Fourier      | Continua, periódica              | Discreto                  | Periódica             | Discreta             |
| TF                    | Continua, no periódica           | Continua                  | No periódica          | Continua             |
| DTFT                  | Discreta, no periódica (infinita)| Continua y periódica      | No periódica          | Periódica (2π)       |
| DFT                   | Discreta, finita (N muestras)    | Discreta y periódica      | Finita                | Discreta y periódica |

---

Estas herramientas permiten analizar señales en diferentes dominios y contextos. La elección adecuada depende del tipo de señal y del análisis que se desea realizar: análisis teórico (TF, DTFT) o computacional (DFT).

### D. ¿En qué consiste el algoritmo FFT y cuál es su utilidad?

####  ¿Qué es la FFT?

La Fast Fourier Transform (FFT) es un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT) de una señal. La DFT convierte una secuencia de valores en el dominio del tiempo en una representación en el dominio de la frecuencia.

El cálculo directo de la DFT tiene una complejidad computacional de orden:

$$
\mathcal{O}(N^2)
$$

donde N es el número de muestras. Esto se debe a que se requieren N multiplicaciones complejas para cada una de las N frecuencias.

La FFT reduce drásticamente esta complejidad a:

$$
\mathcal{O}(N \log_2 N)
$$

Este aumento de eficiencia se logra aprovechando simetrías y redundancias en los cálculos de la DFT, especialmente cuando N es una potencia de 2.

---

####  ¿Cómo funciona?

El algoritmo FFT más conocido es el de Cooley-Tukey, que descompone la DFT de tamaño N en DFTs más pequeñas recursivamente, hasta llegar a DFTs de tamaño 2 (caso base).

Por ejemplo, para una señal de longitud N = 8:

- La señal se divide en dos grupos: índices pares e impares.
- Se calcula la FFT para cada grupo recursivamente.
- Luego, se combinan los resultados usando factores de forma conocidos como "twiddle factors" (raíces de la unidad).

Este enfoque divide el problema grande en problemas más pequeños (divide y vencerás), lo que reduce considerablemente el número total de operaciones.

---

####  Utilidad de la FFT

- Permite el análisis espectral en tiempo real o casi real.
- Es la base de aplicaciones como:
  - Procesamiento de audio y música.
  - Compresión de señales.
  - Detección de fallos en sistemas eléctricos.
  - Reconocimiento de patrones y voz.
  - Análisis en comunicaciones (OFDM, modulación digital).

---

####  Comparación de costos computacionales

| Método | Número de operaciones complejas | Complejidad |
|--------|------------------------------|-------------|
| DFT directa | N²                       | 𝒪(N²)       |
| FFT (Cooley-Tukey) | N log₂ N         | 𝒪(N log₂ N) |

Para N = 1024:

- DFT requiere más de 1 millón de operaciones.
- FFT reduce ese número a aproximadamente 10,240.

Esto permite que transformadas grandes sean computacionalmente viables y rápidas, incluso en dispositivos embebidos.



### E. Explicación detallada del funcionamiento de la FFT y comparación de su costo computacional con la DFT

####  Funcionamiento del algoritmo FFT

La Fast Fourier Transform (FFT) es un algoritmo que calcula de forma eficiente la Transformada Discreta de Fourier (DFT). Su funcionamiento se basa en descomponer la DFT de tamaño N en subproblemas más pequeños, aprovechando simetrías y redundancias matemáticas.

Uno de los algoritmos más comunes de FFT es el Cooley-Tukey. Este asume que N (longitud de la señal) es una potencia de 2 (aunque hay variantes para otros casos) y realiza lo siguiente:

1. Divide la señal de entrada en dos secuencias:
   - Los elementos en índices pares.
   - Los elementos en índices impares.

2. Calcula recursivamente la FFT de estas dos secuencias de longitud N/2.

3. Usa propiedades de periodicidad y simetría de las raíces de la unidad para combinar los resultados de las FFT menores y reconstruir la FFT total.

Esto se repite hasta que las FFTs sean de tamaño 2 (caso base), lo que permite que el número total de operaciones se reduzca significativamente.

Ejemplo visual (para N = 8):

- Entrada x[n]: 8 muestras.
- Se divide en dos grupos: x[0], x[2], x[4], x[6] y x[1], x[3], x[5], x[7].
- Se calcula FFT de cada grupo (4 muestras).
- Luego se combinan los resultados usando factores twiddle:
  W_N^k = e^{-j 2πk/N}

---

####  Comparación del costo computacional: FFT vs. DFT

| Método            | Número de operaciones | Complejidad computacional |
|------------------|------------------------|----------------------------|
| DFT directa       | N² operaciones complejas | 𝒪(N²)                      |
| FFT (Cooley-Tukey) | N log₂ N operaciones     | 𝒪(N log₂ N)                |

La diferencia se hace notable para valores grandes de N.

Ejemplo para N = 1024:

- DFT directa: 1024² = 1,048,576 operaciones.
- FFT: 1024 × log₂(1024) = 1024 × 10 = 10,240 operaciones.

Esto significa que la FFT es aproximadamente 100 veces más rápida que la DFT directa para N = 1024. Esta eficiencia es clave para aplicaciones en tiempo real, como procesamiento de señales, análisis de espectro, comunicaciones y procesamiento de audio e imagen.

---

Conclusión: la FFT no solo conserva el resultado de la DFT, sino que permite obtenerlo con mucho menor esfuerzo computacional, lo que la convierte en una herramienta esencial en ingeniería y ciencias aplicadas.


### **2. Modulación AM por detección coherente**

En este ejercicio se explora la modulación por amplitud (AM) utilizando detección coherente, una técnica ampliamente usada en comunicaciones analógicas. Se estudian sus fundamentos y aplicaciones. Posteriormente, se implementa en Python un ejemplo ilustrativo en el que se visualizan las señales en el dominio del tiempo y la frecuencia (utilizando la transformada rápida de Fourier unidireccional, rfft). Se analizan dos tipos de señales de mensaje: un pulso rectangular y una señal cosenoidal. El índice de modulación es definido por el usuario, lo que permite observar el impacto de este parámetro en la señal modulada.


### A. ¿Qué es la modulación AM con detección coherente?

La modulación por amplitud (AM, por sus siglas en inglés) es una técnica mediante la cual la amplitud de una señal portadora senoidal se varía de forma proporcional a una señal de mensaje o información. En la detección coherente (también llamada detección síncrona), la señal modulada es demodulada utilizando una señal portadora que está perfectamente sincronizada en frecuencia y fase con la portadora original utilizada en la transmisión.

Esta sincronización permite recuperar con mayor precisión la señal de mensaje original, ya que se evita la distorsión que puede generarse con otros métodos de detección no coherente (como la detección por envolvente).

#### Señal modulada en AM:
La señal AM con modulación en amplitud se puede expresar como:

$$
s(t) = A_c \left[1 + m \cdot x(t)\right] \cdot \cos(2\pi f_c t)
$$

Donde:
- \( A_c \): amplitud de la portadora.
- \( x(t) \): señal de mensaje (normalizada entre -1 y 1).
- \( f_c \): frecuencia de la portadora.
- \( m \): índice de modulación.

Para que la señal esté correctamente modulada, se requiere que \( 0 < m \leq 1 \). Si \( m > 1 \), se produce sobre-modulación.

---

### Aplicaciones de la modulación AM con detección coherente

La detección coherente se utiliza cuando se requiere mayor fidelidad en la recuperación de la señal de mensaje. Algunas aplicaciones comunes incluyen:

- Sistemas de comunicación analógica de alta calidad.
- Comunicaciones en radio AM (en algunos receptores avanzados).
- Comunicaciones en banda base de sistemas embebidos.
- Transmisión de datos mediante modulación digital (por ejemplo, QAM, que requiere detección coherente).
- Aplicaciones en radar y telemetría.

Este tipo de detección requiere que tanto el transmisor como el receptor compartan un reloj común o un sistema de sincronización de portadora.


### B. Ejemplo ilustrativo

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# Parámetros generales
Fs = 10000  # Frecuencia de muestreo
t = np.arange(0, 0.01, 1/Fs)  # Tiempo de 10 ms

# Parámetros de modulación
Fc = 1000         # Frecuencia de la portadora
Ac = 1            # Amplitud de la portadora
indice_mod = 0.8  # Índice de modulación definido por el usuario

# Señal mensaje tipo pulso rectangular
m_pulso = np.where((t >= 0.002) & (t < 0.008), 1.0, 0.0)

# Señal mensaje tipo coseno
fm = 200  # frecuencia del mensaje
m_coseno = np.cos(2 * np.pi * fm * t)

# Función para modulación AM
def am_modular(mensaje, Fc, Fs, Ac, indice_mod):
    m_norm = mensaje / np.max(np.abs(mensaje))  # Normalizar mensaje
    portadora = Ac * np.cos(2 * np.pi * Fc * t)
    y = (1 + indice_mod * m_norm) * portadora
    return y

# Señales moduladas
y_pulso = am_modular(m_pulso, Fc, Fs, Ac, indice_mod)
y_coseno = am_modular(m_coseno, Fc, Fs, Ac, indice_mod)

# Gráficas en el tiempo
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(t, m_pulso)
plt.title('Mensaje: Pulso rectangular')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.subplot(2, 2, 2)
plt.plot(t, y_pulso)
plt.title('AM modulada (pulso)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.subplot(2, 2, 3)
plt.plot(t, m_coseno)
plt.title('Mensaje: Coseno')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.subplot(2, 2, 4)
plt.plot(t, y_coseno)
plt.title('AM modulada (coseno)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()

# Espectros usando rfft
def graficar_espectro(signal, Fs, title):
    N = len(signal)
    freqs = rfftfreq(N, 1/Fs)
    spectrum = np.abs(rfft(signal)) / N
    plt.plot(freqs, spectrum)
    plt.title(title)
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Magnitud')

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
graficar_espectro(y_pulso, Fs, 'Espectro AM (pulso)')

plt.subplot(1, 2, 2)
graficar_espectro(y_coseno, Fs, 'Espectro AM (coseno)')

plt.tight_layout()
plt.show()
````

### Visualización de resultados: señales en el tiempo y frecuencia

A continuación se explican las gráficas generadas para los dos tipos de señal mensaje utilizados en la modulación AM:

#### Señales en el dominio del tiempo:

- En la primera fila se presentan las señales en el tiempo:
  - Gráfica 1 (izquierda): señal mensaje tipo pulso rectangular. Es una señal que toma el valor 1 entre 2 ms y 8 ms, y 0 en el resto del intervalo.
  - Gráfica 2 (derecha): señal modulada AM correspondiente al mensaje tipo pulso rectangular. Se observa cómo la envolvente de la portadora varía con el contorno del pulso rectangular.

- En la segunda fila:
  - Gráfica 3 (izquierda): señal mensaje tipo coseno. Es una onda senoidal de baja frecuencia (200 Hz) usada como mensaje.
  - Gráfica 4 (derecha): señal modulada AM correspondiente al mensaje coseno. Se nota cómo la amplitud de la portadora varía siguiendo la forma del coseno (es decir, la envolvente reproduce la señal mensaje).

#### Señales en el dominio de la frecuencia (rfft):

- Gráfica 5 (izquierda): espectro de la señal AM con mensaje tipo pulso rectangular. Se observan múltiples componentes frecuenciales debido a la naturaleza no periódica del pulso. El espectro es más ancho comparado con el caso del coseno.

- Gráfica 6 (derecha): espectro de la señal AM con mensaje tipo coseno. Se observan tres componentes principales: la frecuencia de la portadora (Fc), y dos bandas laterales simétricas ubicadas en Fc ± fm. Esta es la característica típica del espectro de una señal AM de banda doble con portadora.

Estas gráficas permiten visualizar cómo la señal mensaje afecta tanto la forma temporal como la distribución de energía espectral de la señal modulada.

Cortar fragmento de 20 a 25 segundos

````python
import numpy as np
import matplotlib.pyplot as plt

xpro = x.copy() #copiar archivos para procesar
ti = 20 #tiempo incio a procesar seg
tf = 25 #tiempo final a procesar seg
xs = xpro[int(ti*fs):int((tf*fs)),:]

tt = np.arange(ti,tf,1/fs) # vector de tiempo
plt.plot(tt,xs)
plt.legend(('canal 1','canal 2'))
plt.xlabel('$t[s]$')
plt.ylabel('$x(t)$')
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Usar solo canal 1
xm = xs[:,0]

# Normalizar señal de mensaje
xm = xm / max(abs(xm))

# Parámetros de modulación
Fc = 10000 # frecuencia de portadora
Im = 1     # índice de modulación

Am = 1     # amplitud del mensaje
Fs = 5*Fc  # frecuencia de muestreo deseada
ti = 0
tf = 5     # duración en segundos

# Interpolación para nuevo vector de tiempo
t = np.arange(ti, tf, 1/Fs) # vector del tiempo re-muestreado
m = np.interp(t, np.linspace(0, tf, len(xm)), xm) # mensaje re-muestreado

Ac = max(abs(m))/Im # amplitud de la portadora según índice de modulación
c = Ac*np.sin(2*np.pi*Fc*t) # señal portadora

# === GRAFICA EN TIEMPO: MENSAJE Y PORTADORA ===
plt.plot(t, c, label='portadora')
plt.plot(t, m, label='mensaje')
plt.legend()
plt.xlabel('$t[s]$')
plt.ylabel('amplitud')
plt.title('Señales en el tiempo: mensaje (fragmento canción) y portadora')
plt.show()
````

````python
# === SEÑAL MODULADA EN AM ===
y = (1 + m/Ac)*c # señal modulada

plt.plot(t, y)
plt.xlabel('$t[s]$')
plt.ylabel('$y(t)$')
plt.title('Señal modulada en AM del fragmento de canción')
plt.show()
````

````python
# === ESPECTROS DE LAS SEÑALES ===
Xfm = np.fft.fft(m)
Xfc = np.fft.fft(c)
Xfy = np.fft.fft(y)

vfre = np.fft.fftfreq(len(t), 1/Fs)

# Tomar solo la mitad positiva del espectro
half = len(vfre) // 2
vfre_pos = vfre[:half]
Xfm_pos = np.abs(Xfm[:half])
Xfc_pos = np.abs(Xfc[:half])
Xfy_pos = np.abs(Xfy[:half])

# Normalización para mejor visualización
Xfm_pos = Xfm_pos / np.max(Xfm_pos)
Xfc_pos = Xfc_pos / np.max(Xfc_pos)
Xfy_pos = Xfy_pos / np.max(Xfy_pos)

# Graficar espectros
plt.plot(vfre_pos, Xfm_pos, label='mensaje', color='darkorange')
plt.plot(vfre_pos, Xfc_pos, label='portadora', color='firebrick')
plt.plot(vfre_pos, Xfy_pos, label='modulado', color='purple')

plt.xlabel('$f\ [Hz]$')
plt.ylabel('$|X(f)|$')
plt.title('Espectros: mensaje, portadora y señal modulada')

# Leyenda sin duplicados
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.ylim([0, 1.1])  # Ya está normalizado
plt.grid(True)
plt.show()
````

### a) ¿En qué consiste la Distorsión Total de Armónicos (THD)?

La Distorsión Total de Armónicos (THD) es una medida que cuantifica el nivel de distorsión de una señal periódica debido a la presencia de armónicos. Los armónicos son componentes de frecuencia que son múltiplos enteros de la frecuencia fundamental de una señal.

En los sistemas eléctricos y electrónicos, especialmente en circuitos con componentes no lineales (como rectificadores, inversores o cargas electrónicas), las señales de corriente o voltaje pueden distorsionarse y alejarse de una forma senoidal pura. Esta distorsión introduce componentes armónicas que degradan la calidad de la energía eléctrica y pueden afectar el rendimiento de los equipos.

El THD se define como el cociente entre la raíz cuadrada de la suma de los cuadrados de todas las componentes armónicas (excepto la fundamental) y la amplitud de la componente fundamental. Se expresa generalmente en porcentaje (%).

$$
\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

Donde:

$$
V_1 = \text{Amplitud de la componente fundamental}
$$

$$
V_2, V_3, \dots, V_n = \text{Amplitudes de los armónicos de orden 2 al } n
$$

Una **THD baja** indica que la señal está cerca de una forma senoidal pura. En cambio, una **THD alta** revela una fuerte presencia de armónicos y, por tanto, mayor distorsión.

### b) ¿Qué es el factor de potencia en un circuito eléctrico?

El **factor de potencia (FP)** es una medida que describe cuán eficientemente se utiliza la energía eléctrica en un sistema. Representa la relación entre la **potencia activa** (realmente utilizada para realizar trabajo útil) y la **potencia aparente** (total suministrada por la fuente).

Se define como:

$$
\text{FP} = \frac{P}{S}
$$

donde:

- \( P \): Potencia activa (o real), medida en watts (W).  
- \( S \): Potencia aparente, medida en volt-amperios (VA).

El factor de potencia puede expresarse como un valor entre 0 y 1, o como un porcentaje. Un **FP = 1** (o 100%) indica que toda la potencia suministrada se está utilizando eficazmente. Un **FP bajo** significa que una parte significativa de la potencia está siendo desperdiciada en forma de potencia reactiva.

En sistemas sinusoidales, el FP también se puede interpretar como el **coseno del ángulo de desfase \( \phi \)** entre la corriente y el voltaje:

$$
\text{FP} = \cos(\phi)
$$

- Si el FP es **inductivo**, la corriente **retrasa** al voltaje.
- Si es **capacitivo**, la corriente **adelanta** al voltaje.

Un bajo factor de potencia puede aumentar las pérdidas en el sistema y reducir la capacidad de transmisión, por lo que es deseable mantenerlo lo más alto posible.


### c) ¿Cómo puede calcularse el THD usando la Transformada Rápida de Fourier (FFT)?

La Transformada Rápida de Fourier (FFT) permite descomponer una señal periódica (como una corriente o voltaje en un circuito) en una suma de componentes senoidales (armónicos), cada una con una frecuencia, amplitud y fase específicas.

La Distorsión Total Armónica (THD) puede calcularse utilizando los coeficientes de la FFT de la señal. El procedimiento general es:

1. Se calcula la FFT de la señal de interés (por ejemplo, la corriente del circuito).
2. Se obtiene el módulo (magnitud) de la FFT, que corresponde a las amplitudes de cada componente espectral.
3. Se identifica la componente de frecuencia fundamental (la de menor frecuencia distinta de cero).
4. Se suman los cuadrados de las amplitudes de los armónicos (todas las frecuencias múltiples de la fundamental, excluyendo la fundamental).
5. Finalmente, se calcula el THD con la fórmula:

$$
\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

donde:
$$
V_1 = \text{Amplitud de la componente fundamental}
$$

$$
V_2, V_3, \dots, V_n = \text{Amplitudes de los armónicos de orden 2 al } n
$$

### d) ¿Cómo se relaciona el THD con la distorsión del factor de potencia?

El factor de potencia (FP) en sistemas con señales no sinusoidales se ve afectado por dos tipos de distorsiones:

1. El desfase entre la tensión y la corriente (factor de potencia de desplazamiento).
2. La presencia de armónicos en la señal de corriente (factor de potencia de distorsión).

En este contexto, el THD (Distorsión Total Armónica) de la corriente se relaciona directamente con el segundo factor: la distorsión armónica.

El factor de potencia total se puede calcular como:

$$
\text{FP} = \frac{P}{S} = \frac{V_{\text{rms}} I_1 \cos(\phi)}{V_{\text{rms}} I_{\text{rms}}} = \frac{I_1 \cos(\phi)}{I_{\text{rms}}}
$$

Donde:

$$
\begin{aligned}
I_1 & : \text{Componente fundamental de la corriente (amplitud de la FFT en la frecuencia fundamental)} \\
I_{\text{rms}} & : \text{Valor eficaz total de la corriente (incluye fundamental y armónicos)} \\
\phi & : \text{Ángulo de desfase entre la tensión y la corriente fundamental}
\end{aligned}
$$

Ahora bien, la relación entre el THD y el FP se da a través del factor de potencia de distorsión, dado por:

$$
\text{FP}_{\text{distorsión}} = \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

De esta forma, el factor de potencia total se puede expresar como:

$$
\text{FP} = \text{FP}_{\text{desplazamiento}} \times \text{FP}_{\text{distorsión}} = \cos(\phi) \times \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

Esto significa que incluso si el desfase entre tensión y corriente fundamental es pequeño (es decir, \( \cos(\phi) \approx 1 \)), un alto contenido de armónicos (alto THD) reducirá significativamente el factor de potencia total.


e) Diseñar y simular un rectificador de onda completa en dos casos:

I. Carga netamente resistiva

II. Carga RC en serie:

**I. Carga netamente resistiva**

Se va a simular un rectificador de onda completa alimentado por una señal senoidal de 60 Hz con amplitud de 120 V, conectada a una carga puramente resistiva.

**Paso 1: Definir parámetros y generar la señal de entrada**

````python
# Parámetros de simulación
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import scipy.fftpack as fft

# Definición de parámetros
Fo = 60            # Frecuencia de alimentación (Hz)
Fs = 30 * Fo       # Frecuencia de muestreo
To = 1 / Fo        # Periodo fundamental
Ts = 1 / Fs        # Periodo de muestreo
t = np.arange(0, 5 * To, Ts)  # tiempo para 5 periodos

A = 120  # Amplitud de entrada (voltaje)
in_o = A * np.sin(2 * np.pi * Fo * t)  # señal senoidal de entrada
rec_c = sig.square(2 * np.pi * Fo * t)  # tren de pulsos para rectificación completa

# Señal rectificada completamente
in_ = in_o * rec_c

# Graficar señal de entrada rectificada
plt.plot(t, in_, label='Entrada rectificada')
plt.xlabel('t [s]')
plt.ylabel('Voltaje [V]')
plt.title('Entrada al circuito: Rectificador onda completa')
plt.legend()
plt.grid()
plt.show()
````

**Paso 2: Calcular y graficar el espectro para calcular el THD**

````python
# FFT de la señal rectificada
Xf = np.fft.rfft(in_)
vfre = np.fft.rfftfreq(len(in_), 1/Fs)

# Graficar espectro
plt.stem(vfre, np.abs(Xf)/len(in_))  # sin use_line_collection
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de la corriente rectificada')
plt.grid()
plt.show()
````

**Paso 3: Cálculo del THD y del factor de potencia por distorsión**

````python
# THD desde componentes espectrales
Xmag = np.abs(Xf)/len(in_)
fundamental = Xmag[np.argmax((vfre >= Fo) & (vfre <= Fo+1))]  # componente en 60 Hz
harmonics = Xmag[(vfre > Fo + 1)]  # armónicos superiores (excluye la fundamental)

# Calcular THD
THD = np.sqrt(np.sum(harmonics**2)) / fundamental
print(f"THD: {THD:.4f} ({THD*100:.2f}%)")

# Calcular factor de potencia por distorsión
FP_dist = 1 / np.sqrt(1 + THD**2)
print(f"Factor de potencia (por distorsión): {FP_dist:.4f}")
````

**II. Carga RC en serie:**

````python
# Parámetros del circuito
f = 60              # Frecuencia de red [Hz]
w = 2*np.pi*f       # Frecuencia angular [rad/s]
Vm = 170            # Voltaje pico (120 Vrms ≈ 170 V pico)
R = 100             # Resistencia [Ohm]
C = 100e-6          # Capacitancia [F]

T = 1/f             # Periodo de la señal
tf = 4*T            # Tiempo total de simulación
n = 1000            # Puntos por ciclo
t = np.linspace(0, tf, int(n*tf/T))
dt = t[1] - t[0]

# Voltaje de entrada (onda completa rectificada)
v_in = Vm*np.abs(np.sin(w*t))

# Inicialización de variables
i_rc = np.zeros(len(t))
v_c = np.zeros(len(t))

# Simulación de corriente y voltaje en el capacitor
for k in range(1, len(t)):
    di = (v_in[k] - R*i_rc[k-1] - v_c[k-1]) / R
    i_rc[k] = i_rc[k-1] + di*dt
    v_c[k] = v_c[k-1] + i_rc[k]*dt/C
````

### Forma de onda de corriente y voltaje en el condensador

````python
plt.figure(figsize=(10,5))
plt.plot(t, v_in, label='$v_{in}(t)$', linestyle='--', color='gray')
plt.plot(t, i_rc, label='$i_{RC}(t)$', color='green')
plt.plot(t, v_c, label='$v_C(t)$', color='red')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Rectificador de onda completa con carga RC en serie')
plt.legend()
plt.grid()
plt.show()
````

### Espectro de la corriente con carga RC


````python
from scipy.fft import fft, fftfreq

Xf_rc = fft(i_rc)
vfre_rc = fftfreq(len(i_rc), dt)

plt.figure(figsize=(10,4))
plt.stem(vfre_rc, np.abs(Xf_rc)/len(i_rc))
plt.xlim(0, 1000)
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de $i_{RC}(t)$')
plt.grid()
plt.show()
````

### Cálculo del THD y del factor de potencia por distorsión

````python
# Magnitud del espectro
mag_rc = np.abs(Xf_rc[:len(Xf_rc)//2]) / len(i_rc)

# Componente fundamental (60 Hz)
V1_rc = mag_rc[np.argmin(np.abs(vfre_rc[:len(mag_rc)] - f))]

# Armónicos relevantes
harmonicos = [2*f, 3*f, 4*f, 5*f, 6*f]
sum_h_rc = 0
for h in harmonicos:
    idx = np.argmin(np.abs(vfre_rc[:len(mag_rc)] - h))
    sum_h_rc += mag_rc[idx]**2

# THD
THD_rc = np.sqrt(sum_h_rc) / V1_rc
print(f"THD (RC): {THD_rc:.4f} ({THD_rc*100:.2f}%)")

# Factor de potencia por distorsión
FPd_rc = 1 / np.sqrt(1 + THD_rc**2)
print(f"Factor de potencia por distorsión (RC): {FPd_rc:.4f}")
````

### Conclusión

Se observa que la presencia del condensador modifica la forma de onda de la corriente, introduciendo componentes adicionales en el espectro (más armónicos). Esto provoca un incremento en el THD y una disminución en el factor de potencia por distorsión en comparación con la carga puramente resistiva.

Este análisis es fundamental para entender cómo la naturaleza de la carga afecta la calidad de la corriente consumida y su impacto en la eficiencia energética del sistema.
